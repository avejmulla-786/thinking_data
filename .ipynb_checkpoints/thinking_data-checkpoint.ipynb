```python
import json
```

```python
#Load data 
def load_data(filename):
    with open (filename, "r")as f:
        data=json.load(f)

    return data
```

```python
data=load_data("store_data.json")
```

```python
print(data)
```

```python
def clean_data(data):
    text_to_num={"one":1,"two":2,"three":3,"four":4,"five":5}
    cleaned_data = []
    unique_users = set()
    for user in data:
        row_rating=str(user["rating"]).strip().lower()
        if(row_rating in text_to_num):
            row_rating = text_to_num[row_rating]
        
        user['rating']=row_rating

        row_age = user.get("age")
        if (row_age == None):
            user["age"] = None  

        if(user["name"].strip()in unique_users):
            continue 
        unique_users.add(user["name"])    
        cleaned_data.append(user)
        
    return cleaned_data                     
```

```python
data=clean_data(data)
```

```python
def get_insights(data):

    # avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])

    print(f"avg rating = {tot_rating/len(data)}")

    # percentage of users with poor rating (< 3)
    poor = 0
    for user in data:
        if float(user["rating"]) < 3:
            poor += 1

    percent_poor = (poor / len(data)) * 100
    print(f"percentage of poor rating users = {percent_poor}%")
```

```python
get_insights(data)
```

```python
#recomendation of user
def get_recomendation(data):
    recomendations = []

    for user in data:
        curr_recomn = {}
        curr_recomn["name"] = user["name"]

        if(float(user["rating"]) >= 4):
            curr_recomn["brand"] = "apple"

        else:
            curr_recomn["brand"] = "samsang"

        recomendations.append(curr_recomn)
    return recomendations 
```

```python
get_recomendation(data)
```
